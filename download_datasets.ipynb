```
# SCENES
import os
import numpy as np
import random

!wget http://groups.csail.mit.edu/vision/LabelMe/NewImages/indoorCVPR_09.tar
!wget http://web.mit.edu/torralba/www/TrainImages.txt
!wget http://web.mit.edu/torralba/www/TestImages.txt
!tar -xf /content/indoorCVPR_09.tar

input_dir = './'
dataset = 'scenes'
n_classes = 67

os.mkdir(os.path.join(input_dir, dataset))
labels = []

for f in os.listdir(os.path.join(input_dir, 'Images')):
    if not f.startswith('.'):
        labels.append(f)

with open(os.path.join(input_dir, 'TrainImages.txt')) as f:
    train_files = f.readlines()
    for train_file in train_files:
        print(train_file.replace('\n', ''))
        labelname = (train_file.replace('\n', '')).split('/')[0]
        filename = (train_file.replace('\n', '')).split('/')[1]

        label_id = None
        for i, label in enumerate(labels):
            if label == labelname:
                label_id = i
                break
        src = os.path.join(input_dir, 'Images', train_file.replace('\n', ''))
        dst = os.path.join(input_dir, dataset, str(label_id) + '_' + filename.replace('\n', ''))
        os.system('mv ' + src + ' ' + dst)

with open(os.path.join(input_dir, 'TestImages.txt')) as f:
    test_files = f.readlines()
    for test_file in test_files:
        labelname = test_file.split('/')[0]
        filename = test_file.split('/')[1]

        label_id = None
        for i, label in enumerate(labels):
            if label == labelname:
                label_id = i
                break
        src = os.path.join(input_dir, 'Images', test_file.replace('\n', ''))
        dst = os.path.join(input_dir, dataset, str(label_id) + '_' + filename.replace('\n', ''))
        os.system('mv ' + src + ' ' + dst)
        
'''!rm indoorCVPR_09.tar TestImages.txt TrainImages.txt
!rm -rf Images'''
```

```
# BIRDS
import os
import numpy as np
import random

!wget http://www.vision.caltech.edu/visipedia-data/CUB-200-2011/CUB_200_2011.tgz
!tar -xf CUB_200_2011.tgz

input_dir = './'
dataset = 'birds'
n_classes = 200

os.mkdir(os.path.join(input_dir, dataset))
with open(os.path.join(input_dir, 'CUB_200_2011', 'images.txt')) as f:
    files_list = f.readlines()

for f in files_list:
    f = f.replace('\n', '').split(' ')[1]
    y = int(f.split('.')[0]) - 1
    img_id = f.split('_')[-1]
    src = os.path.join(input_dir, 'CUB_200_2011', 'images', f)
    dst = os.path.join(input_dir, dataset, str(y) + '_' + img_id)
    os.system('mv ' + src + ' ' + dst)
        
'''!rm *.tgz
!rm -rf lists images CUB*'''
```

```
# FLOWERS
import os
import numpy as np
from scipy.io import loadmat
import random

!wget http://www.robots.ox.ac.uk/~vgg/data/flowers/102/102flowers.tgz
!wget http://www.robots.ox.ac.uk/~vgg/data/flowers/102/imagelabels.mat
!tar -xf 102flowers.tgz

input_dir = './'
dataset = 'flowers'
n_classes = 102

os.mkdir(os.path.join(input_dir, dataset))

labels_mat = loadmat(os.path.join(input_dir, 'imagelabels.mat'))
labels = labels_mat['labels'][0]

for f in os.listdir(os.path.join(input_dir, 'jpg')):
    if not f.startswith('.'):
        f = f.replace('\n', '')
        img_id = int((f.split('_')[1]).split('.')[0])
        y = labels[img_id - 1] - 1
        src = os.path.join(input_dir, 'jpg', f)
        dst = os.path.join(input_dir, dataset, str(y) + '_' + str(img_id) + '.jpg')
        os.system('mv ' + src + ' ' + dst)
        
'''!rm *.tgz *.mat
!rm -rf jpg'''
```

```
# CARS
import os
import numpy as np
from scipy.io import loadmat
import random

!wget http://imagenet.stanford.edu/internal/car196/car_ims.tgz
!wget http://imagenet.stanford.edu/internal/car196/cars_annos.mat
!tar -xf car_ims.tgz

input_dir = './'
dataset = 'cars'
n_classes = 196

os.mkdir(os.path.join(input_dir, dataset))

cars_annos_mat = loadmat(os.path.join(input_dir, 'cars_annos.mat'))
annotations = cars_annos_mat['annotations'][0]

for annotation in annotations:
    file_name = annotation[0][0].split('/')[1]
    y = int(annotation[-2][0][0]) - 1
    src = os.path.join(input_dir, 'car_ims', file_name)
    dst = os.path.join(input_dir, dataset, str(y) + '_' + file_name)
    os.system('mv ' + src + ' ' + dst)

'''!rm *.tgz *.mat
!rm -rf car_ims'''
```

```
# AIRCRAFTS
import os
import numpy as np
from scipy.io import loadmat
import random
from PIL import Image

!wget http://www.robots.ox.ac.uk/~vgg/data/fgvc-aircraft/archives/fgvc-aircraft-2013b.tar.gz
!tar -xf /content/fgvc-aircraft-2013b.tar.gz

input_dir = './'
dataset = 'aircrafts'
n_classes = 90

os.mkdir(os.path.join(input_dir, dataset))

lines = os.listdir(os.path.join(input_dir, 'fgvc-aircraft-2013b/data', 'images'))
files = []
for line in lines:
    files.append((line.replace('\n', '')).split('.')[0])

labels = []
with open(os.path.join(input_dir, 'fgvc-aircraft-2013b/data', 'images_variant_train.txt')) as f:
    train_files = f.readlines()
    for train_file in train_files:
        label = (train_file.replace('\n', '')).split(' ')[1]
        labels.append(label)
    labels = np.unique(labels)

    for train_file in train_files:
        file_id = str((train_file.replace('\n', '')).split(' ')[0])
        label = (train_file.replace('\n', '')).split(' ')[1]
        for i in range(len(labels)):
            if labels[i] == label:
                label_id = str(i)
        src = os.path.join(input_dir, 'fgvc-aircraft-2013b/data', 'images', file_id + '.jpg')
        dst = os.path.join(input_dir, dataset, label_id + '_' + file_id + '.jpg')
        os.system('mv ' + src + ' ' + dst)

with open(os.path.join(input_dir, 'fgvc-aircraft-2013b/data', 'images_variant_val.txt')) as f:
    val_files = f.readlines()
    for val_file in val_files:
        file_id = str((val_file.replace('\n', '')).split(' ')[0])
        label = (val_file.replace('\n', '')).split(' ')[1]
        for i in range(len(labels)):
            if labels[i] == label:
                label_id = str(i)
        src = os.path.join(input_dir, 'fgvc-aircraft-2013b/data', 'images', file_id + '.jpg')
        dst = os.path.join(input_dir, dataset, label_id + '_' + file_id + '.jpg')
        os.system('mv ' + src + ' ' + dst)

with open(os.path.join(input_dir, 'fgvc-aircraft-2013b/data', 'images_variant_test.txt')) as f:
    test_files = f.readlines()
    for test_file in test_files:
        file_id = str((test_file.replace('\n', '')).split(' ')[0])
        label = (test_file.replace('\n', '')).split(' ')[1]
        for i in range(len(labels)):
            if labels[i] == label:
                label_id = str(i)
        src = os.path.join(input_dir, 'fgvc-aircraft-2013b/data', 'images', file_id + '.jpg')
        dst = os.path.join(input_dir, dataset, label_id + '_' + file_id + '.jpg')
        os.system('mv ' + src + ' ' + dst)

for f in os.listdir(os.path.join(input_dir, dataset)):
    im = np.asarray(Image.open(os.path.join(input_dir, dataset, f)))
    adjusted_im = Image.fromarray(im[:-15, :, :])
    adjusted_im.save(os.path.join(input_dir, dataset, f))

        
'''!rm *.gz
!rm -rf fgvc*'''
```
